```python
import cv2
```

```python
#Uploading image
image=cv2.imread("/content/maxresdefault.jpg")
```

```python
#Importing necesaary libraries
#converting the image into RGB format
import matplotlib.pyplot as plt
%matplotlib inline
img_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#Converting the image into greyscale
grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
grey_image
```

```python
plt.imshow(grey_image)
```

```python
#Inverted Image
inverted_img=255-grey_image
plt.imshow(inverted_img)
```

```python
blurred=cv2.GaussianBlur(inverted_img,(21,21),0)
plt.imshow(blurred)
```

```python
inverted_blurred=255-blurred
pencil_sketch=cv2.divide(grey_image,inverted_blurred,scale=256.0)
plt.imshow(pencil_sketch)
```

```python
#The final sketch
from google.colab.patches import cv2_imshow
cv2_imshow(pencil_sketch)
```
